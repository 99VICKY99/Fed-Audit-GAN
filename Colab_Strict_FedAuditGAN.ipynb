```
# Plot results
import pickle
import matplotlib.pyplot as plt

with open('./results_strict/strict_history.pkl', 'rb') as f:
    history = pickle.load(f)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(history['test_accuracy'], 'b-o')
axes[0].set_title('Test Accuracy')
axes[0].set_xlabel('Round')
axes[1].plot(history['baseline_bias'], 'r-s')
axes[1].set_title('Baseline Bias')
axes[1].set_xlabel('Round')
plt.tight_layout()
plt.show()
```
